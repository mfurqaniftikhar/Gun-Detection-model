```python
model_1 = YOLO('yolo.pt')
```

```python
import gradio as gr
import numpy as np
```

```python
from PIL import Image
def object_detection(image):
  pred = model_1.predict(image)
  pec_image=pred[0].plot()
  return Image.fromarray(pec_image)
```

```python
app = gr.Interface(fn=object_detection,
                   inputs='image',
                   outputs='image'
)
```

```python
app.launch(show_error=True)
```

**GUN OBJECT DETECTION**

```python
import zipfile
```

```python
zip = zipfile.ZipFile('/content/gun_data.v2i.yolov11.zip')
```

```python
zip.extractall()
```

```python
from ultralytics import YOLO
model = YOLO('yolo11n.pt')
```

```python
model.train(data='/content/data.yaml', epochs=10, imgsz=640)
```

```python
model.save('gun_detection_model.pt')
```

```python
gun_model = YOLO('gun_detection_model.pt')
```

```python
from IPython import display
display.Image('/content/gun.jpeg')
```

```python
pred =  gun_model.predict('/content/gun.jpeg')
```

```python
pred[0].plot()
```

```python
def gun_detection(image):
  pred = gun_model.predict(image)
  pec_image=pred[0].plot()
  return Image.fromarray(pec_image)
```

```python
import gradio as gr
```

```python
appp = gr.Interface(fn=gun_detection,
                   inputs='image',
                   outputs='image'
)
```

```python
appp.launch(show_error=True)
```
